```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import pyautogui
import time

#Creating a variable ----------------
wordToSearch = "playstation 4"

#Opening Browser ----------------
driver = webdriver.Chrome()
driver.maximize_window()

#Website 1 ----------------
driver.get("https://www.mercadolivre.com.br/")
searchBar = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="cb1-edit"]')))
searchBar.click()
searchBar.send_keys(wordToSearch)
searchIcon = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/header/div/form/button')))
searchIcon.click()

#Website 2 ----------------
time.sleep(1)
driver.switch_to.new_window('tab')
driver.get("https://www.amazon.com.br/")
searchBar = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="twotabsearchtextbox"]')))
searchBar.click()
searchBar.send_keys(wordToSearch)
searchIcon = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="nav-search-submit-button"]')))
searchIcon.click()

#Website 3 ----------------
time.sleep(1)
driver.switch_to.new_window('tab')
driver.get("https://www.ebay.com/")
searchBar = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="gh-ac"]')))
searchBar.click()
searchBar.send_keys(wordToSearch)
searchIcon = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="gh-btn"]')))
searchIcon.click()

#Website 4 ----------------
time.sleep(1)
driver.switch_to.new_window('tab')
driver.get("https://www.magazineluiza.com.br/")
searchBar = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="input-search"]')))
searchBar.click()
searchBar.send_keys(wordToSearch)
searchBar.send_keys(Keys.ENTER)

#Website 5 ----------------
time.sleep(1)
driver.switch_to.new_window('tab')
driver.get("https://www.olx.com.br/")
searchBar = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="searchtext"]')))
searchBar.click()
searchBar.send_keys(wordToSearch)
searchBar.send_keys(Keys.ENTER)

#Website 6 ----------------
#time.sleep(1)
#driver.switch_to.new_window('tab')
#driver.get("https://www.americanas.com.br/")
#searchBar = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="rsyswpsdk"]/div/header/div[1]/div[1]/div/div[1]/form/input')))
#searchBar.click()
#searchBar.send_keys(wordToSearch)
#searchIcon = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="rsyswpsdk"]/div/header/div[1]/div[1]/div/div[1]/form/button')))
#searchIcon.click()
```
